Filtro de média (LPF) no domínio do espaço

Importar bibliotecas cv, numpy e matplotlib

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

Carrega a imagem

```python
img = cv.imread('022.png')
#img = cv.imread('SaltAndPepperNoise.jpg')
#img = cv.imread('cameraman.jpg')
```

desenvolvimento do kernel para filtro de média de dimensões 3, 4 e 5

```python
filt3 = np.ones((3, 3), np.float32) / 9

filt5 = np.ones((5, 5), np.float32) / 25

filt7 = np.ones((7, 7), np.float32) / 49

```

Desenvolvimento de filtro com kernel personalizado

```python
filt16 = np.array([[1,2,1],
                  [2,4,2],
                  [1,2,1]]) / 16
```

Aplicação dos filtros na imagem

```python
res3 = cv.filter2D(img,-1,filt3)

res5 = cv.filter2D(img,-1,filt5)

res7 = cv.filter2D(img,-1,filt7)

res16 = cv.filter2D(img,-1,filt16)
```

Aplicação dos filtros de mediana de tamanhos 3, 5 e 7

```python
median3 = cv.medianBlur(img,5)

median5 = cv.medianBlur(img,5)

median7 = cv.medianBlur(img,5)
```

plot das imagens

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(res3),plt.title('Media 3')
plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(121),plt.imshow(res5),plt.title('Media 5')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(res7),plt.title('Media 7')
plt.xticks([]), plt.yticks([])
plt.show()
```
